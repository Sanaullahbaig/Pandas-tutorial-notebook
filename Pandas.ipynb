**Series** -> It is one dimensional array that is capable of storing various data types.

```python
import pandas as pd
x = [6, 7, 8, 9]
a = pd.Series(x)
print(a)
print(type(a))
# get value by index number
print(a[2])

# parameters ->index will be a,b,c,d and datatype will be float and name will be python
b = pd.Series(x, index=['a', 'b', 'c', 'd'], dtype='float', name='python')
print(b)
```

Dictionary in Series

```python
dic = {
    "name" : ['python', 'C', 'C++', 'Java'],
    "popularity" : [12, 13, 14, 15],
    "rank" : [1, 4, 3, 2]
}

a = pd.Series(dic) # datatype will be object as there is mixed data

print(a)
```

```python
# Multiplication
var['C'] = var['A'] * var['B']
print(var)
```

```python
# Division
var['C'] = var['A'] / var['B']
print(var)
```

Handling Missing Data -- dropna & fillna

**dropna**

```python
csv_11 = pd.read_csv("text.csv")
csv_11.dropna() # It will drop the rows with missing data
```

```python
csv_11.dropna(axis = 1) # drop along column
```

```python
csv_11.dropna(axis = 0) # drop along rows
```

```python
csv_11.dropna(how = "any") # drop all the rows with atleast one NaN value
```

```python
csv_11.dropna(how = "all") # drop the rows with all values NaN, it will not drop the rows which have atleast one value
```

```python
csv_11.dropna(subset = ["Gender"]) # drop the rows which have NaN values in only Gender column
```

```python
csv_11.dropna(thresh = 2) # drop the rows which have the NaN value 2
```

**fillna**

```python
csv_11.fillna("Python") # here give the valuew you want to fill with
```

```python
# fill value in a particular column we can give the data in a dictionary
csv_11.fillna({"Gender" : "Male", "Department": "Computer Science", "City" : "Islamabad", "Marks" : 80})
```

```python
# fill data from previous data or next data, we can also give here axis to fill the data along axis 0 for rows and 1 for column
csv_11.fillna(method = "ffill", axis=0) # it will fill the previous row's data and use bfill to fill data from next row
```

```python
# fill value by giving it
csv_11.fillna(12, inplace=True) # It will fill all the NaN values by 12
csv_11
```

```python
csv_11.fillna("python",   limit = 1) # It will fill the first NaN value of every column with python
```

**Merging And Concat**

**Merge**  merge() joins DataFrames by matching common columns (like SQL join)

```python
# It will only merge the same values, the column name should be same
vari_1 = pd.DataFrame({"A" : [1, 2, 3, 4], "B" : [5, 6, 7, 8]})
vari_2 = pd.DataFrame({"A" : [1, 2, 3, 5], "C" : [15, 16, 17, 18]})
# Two methods to merge
pd.merge(vari_1, vari_2)
# pd.merge(vari_2, vari_1, on="A")
```

```python
# When some values are not same it will not merge
vari_1 = pd.DataFrame({"A" : [1, 2, 3, 4], "B" : [5, 6, 7, 8]})
vari_2 = pd.DataFrame({"A" : [1, 2, 5, 6], "C" : [15, 16, 17, 18]})
# we can also use this method
pd.merge(vari_2, vari_1, on="A")
```

```python
# how, bydefault it's inner, we can use right, left and outer
# right will do it from right side and left will do from left side, and outer for both data
vari_1 = pd.DataFrame({"A" : [1, 2, 3, 4], "B" : [5, 6, 7, 8]})
vari_2 = pd.DataFrame({"A" : [1, 2, 5, 6], "C" : [15, 16, 17, 18]})

pd.merge(vari_1, vari_2, how="left") # it will do from left as there is vari_1 is on left it will do from vari_1
```

```python
vari_1 = pd.DataFrame({"A" : [1, 2, 3, 4], "B" : [5, 6, 7, 8]})
vari_2 = pd.DataFrame({"A" : [1, 2, 5, 6], "C" : [15, 16, 17, 18]})

pd.merge(vari_1, vari_2, how="right") # it will do from right as there is vari_2 is on right it will do from vari_2
```

```python
# Use outer for both data
# indicator -> it's tell the about the merged data
vari_1 = pd.DataFrame({"A" : [1, 2, 3, 4], "B" : [5, 6, 7, 8]})
vari_2 = pd.DataFrame({"A" : [1, 2, 5, 6], "C" : [15, 16, 17, 18]})

pd.merge(vari_1, vari_2, how="outer", indicator=True)
```

```python
# When both columns names are same but all data is not same so it will only display the column name
vari_1 = pd.DataFrame({"A" : [1, 2, 3, 4], "B" : [5, 6, 7, 8]})
vari_2 = pd.DataFrame({"A" : [1, 2, 5, 6], "B" : [15, 16, 17, 18]})

# To sort this we use index
pd.merge(vari_1, vari_2, left_index=True, right_index=True)
```

```python
# To change the x and y to a name in the last we use suffixes
pd.merge(vari_1, vari_2, left_index=True, right_index=True, suffixes=("My", "Your"))
```

**Concat**  concat() simply stacks DataFrames on top of or beside each other.

```python
# concat in Series
sr1 = pd.Series([1, 2, 3, 4])
sr2 = pd.Series([11, 21, 13, 14])

pd.concat([sr1, sr2]) # here we will pass the data as a list
```

```python
# concat in dataFrame
dt_1 = pd.DataFrame({"A" : [1, 2, 3, 4], "B" : [5, 6, 7, 8]})
dt_2 = pd.DataFrame({"A" : [1, 2, 5, 6], "B" : [15, 16, 17, 18]})

pd.concat([dt_1, dt_2])
```

```python
# When column names are not same
dt_3 = pd.DataFrame({"A" : [1, 2, 3, 4], "B" : [5, 6, 7, 8]})
dt_4 = pd.DataFrame({"D" : [1, 2, 5, 6], "C" : [15, 16, 17, 18]})

pd.concat([dt_3, dt_4])
```

```python
# we can merge it accrdig to axis 0 for rows and 1 for column by default it works in rows
dt_5 = pd.DataFrame({"A" : [1, 2, 3, 4], "B" : [5, 6, 7, 8]})
dt_6 = pd.DataFrame({"A" : [1, 2, 5, 6], "B" : [15, 16, 17, 18]})

pd.concat([dt_5, dt_6], axis=1)
```

```python
# as we have missing data in dt_8 it will show NaN, so sort this we will use join
# join has two values inner and outer inner will show intersection and outer will show union/all values
dt_7 = pd.DataFrame({"A" : [1, 2, 3, 4], "B" : [5, 6, 7, 8]})
dt_8 = pd.DataFrame({"A" : [1, 2], "B" : [15, 16]})

pd.concat([dt_7, dt_8], axis=1, join="inner")
```

```python
# giving keys to the merging data
dt_9 = pd.DataFrame({"A" : [1, 2, 3, 4], "B" : [5, 6, 7, 8]})
dt_0 = pd.DataFrame({"A" : [1, 2, 5, 6], "B" : [15, 16, 17, 18]})

pd.concat([dt_9, dt_0], axis=1, keys=["d1", "d2"])
```

```python
# same program in 0 axis
# giving keys to the merging data
dt_9 = pd.DataFrame({"A" : [1, 2, 3, 4], "B" : [5, 6, 7, 8]})
dt_0 = pd.DataFrame({"A" : [1, 2, 5, 6], "B" : [15, 16, 17, 18]})

pd.concat([dt_9, dt_0], axis=0, keys=["d1", "d2"])
```

```python
# one dimension and 2 dimension DataFrame
dt_10 = pd.DataFrame({"A" : [1, 2, 3, 4]})
dt_11 = pd.DataFrame({"A" : [1, 2, 5, 6], "B" : [15, 16, 17, 18]})

pd.concat([dt_10, dt_11])
```

```python
# one dimension and 2 dimension DataFrame, when columns name are different
dt_12 = pd.DataFrame({"C" : [1, 2, 3, 4]})
dt_13 = pd.DataFrame({"A" : [1, 2, 5, 6], "B" : [15, 16, 17, 18]})

pd.concat([dt_12, dt_13])
```

**Join**

```python
ja_1 = pd.DataFrame({"A" : [1, 2, 3, 4], "B" : [5, 6, 7, 8]})
ja_2 = pd.DataFrame({"C" : [11, 22, 13, 14], "D" : [15, 16, 17, 18]})

ja_1.join(ja_2)
```

```python
# when some data is missing
ja_3 = pd.DataFrame({"A" : [1, 2, 3, 4], "B" : [5, 6, 7, 8]})
ja_4 = pd.DataFrame({"C" : [11, 22], "D" : [15, 16]})

ja_3.join(ja_4)
```

```python
# giving index number, we have to give the index number to both otherwise it will show NaN value where we don't give index
ja_5 = pd.DataFrame({"A" : [1, 2, 3, 4], "B" : [5, 6, 7, 8]}, index=['a', 'b', 'c', 'd'])
ja_6 = pd.DataFrame({"C" : [11, 22], "D" : [15, 16]}, index=['a', 'b'])

ja_5.join(ja_6)
```

```python
# when we have the first column with low data and second with high i'll print the data according to first column
# means it will skip the other data whose index number are not in first column

ja_7 = pd.DataFrame({"A" : [1, 2, 3, 4], "B" : [5, 6, 7, 8]}, index=['a', 'b', 'c', 'd'])
ja_8 = pd.DataFrame({"C" : [11, 22], "D" : [15, 16]}, index=['a', 'b'])

ja_8.join(ja_7)
```

```python
# here we use the how to tell from where you have to join
# right for right, left for left, inner for intersection, outer for union
ja_7 = pd.DataFrame({"A" : [1, 2, 3, 4], "B" : [5, 6, 7, 8]}, index=['a', 'b', 'c', 'd'])
ja_8 = pd.DataFrame({"C" : [11, 22], "D" : [15, 16]}, index=['a', 'b'])

ja_8.join(ja_7, how='right')
```

```python
# when column names are same  we can use the lsuffix and rsuffix to change the name
ja_9 = pd.DataFrame({"A" : [1, 2, 3, 4], "B" : [5, 6, 7, 8]}, index=['a', 'b', 'c', 'd'])
ja_0 = pd.DataFrame({"C" : [11, 22], "B" : [15, 16]}, index=['a', 'b'])

ja_9.join(ja_0, how='left', lsuffix= "S", rsuffix="G")
```
